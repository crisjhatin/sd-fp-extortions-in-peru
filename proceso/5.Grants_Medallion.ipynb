##Grants

```python
%sql
-- Otorgar todos los privilegios a catalog a nuevos usuarios
GRANT ALL PRIVILEGES ON CATALOG catalog_cr TO `newuser@gmail.com` ;
GRANT ALL PRIVILEGES ON CATALOG catalog_cr TO `newuser2@gmail.com` 
```

```python
%sql
-- Otorgar permisos de uso a catalog a usuario newuser3
GRANT USE CATALOG ON CATALOG catalog_cr TO `newuser3@gmail.com` ;
```

```python
%sql
-- Otorgar permisos de uso de schemas bronze y silver a newuser3
GRANT USE SCHEMA ON SCHEMA catalog_cr.bronze TO `newuser3@gmail.com` ;
GRANT USE SCHEMA ON SCHEMA catalog_cr.silver TO `newuser3@gmail.com` ;
```

```python
%sql
-- Otorgar permisos de lectura sobre las tablas crime_reports, crimes y months a newuser3
GRANT SELECT ON TABLE catalog_cr.bronze.crime_reports TO `newuser3@gmail.com` ;
GRANT SELECT ON TABLE catalog_cr.bronze.crimes TO `newuser3@gmail.com` ;
GRANT SELECT ON TABLE catalog_cr.bronze.months TO `newuser3@gmail.com` ;
    
-- Otorgar permisos de lectura sobre la tabla crime_reports_transformed a newuser3
GRANT SELECT ON TABLE catalog_cr.silver.crime_reports_transformed TO `newuser3@gmail.com`
```

```python
%sql
-- Mostrar permisos otorgados sobre el catalogo
SHOW GRANTS ON CATALOG catalog_cr;
```
